```python

mean_temps = {}
std_temps = {}

mean_temps_daytime = {}
std_temps_daytime = {}

mean_temps_nighttime = {}
std_temps_nighttime = {}

for name in file_map:
    df = pd.read_csv(
        file_map[name],
        sep=";",
        decimal=",",
        usecols=[0,1,2,3,4],
        header=None,
        names=["ID", "Date", "Time", "Temp", "Humidity"]
    )

    # combineer datum en tijd tot datetime
    df["Datetime"] = pd.to_datetime(df["Date"] + " " + df["Time"], errors="coerce")

    # extract uur uit tijd
    df["Hour"] = df["Datetime"].dt.hour

    # dag = 6 t/m 17 (dus 06:00–18:00)
    daytime = df[(df["Hour"] >= 6) & (df["Hour"] < 18)]
    nighttime = df[(df["Hour"] < 6) | (df["Hour"] >= 18)]

    # berekeningen
    mean_temps[name] = df["Temp"].mean()
    std_temps[name] = df["Temp"].std()
    mean_temps_daytime[name] = daytime["Temp"].mean()
    std_temps_daytime[name] = daytime["Temp"].std()
    mean_temps_nighttime[name] = nighttime["Temp"].mean()
    std_temps_nighttime[name] = nighttime["Temp"].std()

print(" Mean Temperatures (Overall):")
print(mean_temps)
print(" Mean Daytime Temperatures:")
print(mean_temps_daytime)
print(" Mean Nighttime Temperatures:")
print(mean_temps_nighttime)

# --- Plot 1: Overall ---
plt.figure(figsize=(10, 6))
plt.ylabel("Mean Temperature (°C)")
plt.title("Mean Temperatures (Overall) with Standard Deviation")
plt.xticks(rotation=45, ha='right')
plt.bar(mean_temps.keys(), mean_temps.values(),
        yerr=std_temps.values(), capsize=5,  color='skyblue')
plt.tight_layout()
plt.show()

# --- Plot 2: Daytime ---
plt.figure(figsize=(10, 6))
plt.ylabel("Mean Temperature (°C)")
plt.title(" Mean Daytime Temperatures with Standard Deviation")
plt.xticks(rotation=45, ha='right')
plt.bar(mean_temps_daytime.keys(), mean_temps_daytime.values(),
        yerr=std_temps_daytime.values(), capsize=5,  color='orange')
plt.tight_layout()
plt.show()

# --- Plot 3: Nighttime ---
plt.figure(figsize=(10, 6))
plt.ylabel("Mean Temperature (°C)")
plt.title(" Mean Nighttime Temperatures with Standard Deviation")
plt.xticks(rotation=45, ha='right')
plt.bar(mean_temps_nighttime.keys(), mean_temps_nighttime.values(),
        yerr=std_temps_nighttime.values(), capsize=5, color='midnightblue')
plt.tight_layout()
plt.show()
```
